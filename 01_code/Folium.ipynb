```python
endangered = gpd.read_file('Endangered_partial_clean/')
```

```python
africa_endangered = endangered[endangered['region'] == 'Africa']
```

```python
africa_endangered['speakers'] = africa_endangered['speakers'].str.replace('~','')
```

```python
africa_endangered['speakers'] = africa_endangered['speakers'].str.replace('<','')
```

```python
africa_endangered['speakers'] = africa_endangered['speakers'].str.replace('>','')
```

```python
africa_endangered['speakers'] = africa_endangered['speakers'].str.replace('?','')
```

```python
africa_endangered['speakers'] = africa_endangered['speakers'].str.replace(' in Burkina Faso','')
```

```python
africa_endangered['speakers'].fillna(0, inplace=True)
```

```python
africa_endangered['speakers'] = africa_endangered['speakers'].str.replace(',','')
```

```python
africa_endangered['speakers_clean'] = 0
```

```python
for i in africa_endangered['speakers']:
    try:
        africa_endangered['speakers_clean'] = float(i)
    except:
        print(i)
```

```python
africa_endangered
```

```python
slim = africa_endangered.drop(africa_endangered.columns.difference(['lat','long']), 1).dropna()
```

```python
slim = slim.astype(float)
```

```python
slim['weight'] = africa_endangered['speakers_clean'].fillna(0.1)
```

```python
slim_array = slim[['long', 'lat']].values
```

```python
help(folium.Icon)
```

```python
markers = africa_endangered[['lat', 'long','speakers', 'name']].dropna()
```

```python
# Create a Map instance
m = folium.Map(location=[0, 18.35], tiles = 'stamenterrain', zoom_start=10, control_scale=True)

# Add heatmap to map instance
# Available parameters: HeatMap(data, name=None, min_opacity=0.5, max_zoom=18, max_val=1.0, radius=25, blur=15, gradient=None, overlay=True, control=True, show=True)
HeatMap(slim_array).add_to(m)

# Alternative syntax:
# m.add_child(HeatMap(, radius=15))

```

```python

# Add marker
# Run: help(folium.Icon) for more info about icons
for i in markers.index:
    folium.Marker(
        location=[float(markers.long[i]), float(markers.lat[i])],
        popup=markers['name'][i],
        icon=folium.Icon(color='green', icon='glyphicon-asterisk'),
    ).add_to(m)

#Show map
```

```python
m.save('endangered_languages_heatmap.html')
```

```python
guardian = gpd.read_file('../language_data/guardian_language_shp/')
```

```python
[i for i in guardian.columns]
```

```python
markers = guardian[['Longitude', 'Latitude','Name in En', 'Countries','Degree of','Number of']].dropna()
```

```python
points = guardian[['Latitude','Longitude']].dropna().astype(float).values
```

```python
type(points)
```

```python
markers.head(0)
```

```python
markers['Degree of'].value_counts()
```

```python
color_map = {'Extinct': 'black',
             'Severly endangered':'red',
             'Critically endangered' : 'orange',
             'Definitely endangered' : 'yellow',
             'Vulnerable' : 'green'
            }
```

```python
markers['color'] = markers['Degree of'].map(color_map)
```

```python
# Create a Map instance
m = folium.Map(location=[0, 18.35], tiles = 'OpenStreetMap', zoom_start=10, control_scale=True)

# Add heatmap to map instance
# Available parameters: HeatMap(data, name=None, min_opacity=0.5, max_zoom=18, max_val=1.0, radius=25, blur=15, gradient=None, overlay=True, control=True, show=True)
HeatMap(points).add_to(m)

# Alternative syntax:
# m.add_child(HeatMap(, radius=15))


# Add marker
# Run: help(folium.Icon) for more info about icons
for i in markers.index:
    folium.Marker(
        location=[float(markers.Latitude[i]), float(markers.Longitude[i])],
        popup=markers['Name in En'][i],
        icon=folium.Icon(color=markers.color[i], icon='glyphicon-asterisk'),
    ).add_to(m)

#Show map
```

```python
m.save('../maps_and_images/guardian.html')
```
